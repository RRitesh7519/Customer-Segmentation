Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

Data Collection & Processing

```python
#loading the data from csv file into the pandas DataFrames
customer_data = pd.read_csv('/content/Mall_Customers.csv')
```

Choosing the annual income and spending score column

```python
x = customer_data.iloc[:,[3,4]].values
```

```python
print(x)
```

Choosing the number of clusters

WCSS -> Within Cluster Sum of Squares

```python
# finding WCSS value for different number of clusters

wcss=[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  #fir this data to k-means
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
#plot an elbow graph to find which cluster has minimum value
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum number of clusters will be 5

Traing the K-Means clustering Model

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
#return a label for each data point bassed on their cluster
y=kmeans.fit_predict(x)
print(y)
```

visualizing all the clusters

```python
# plotting all the clusters and their centroids

plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0], x[y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(x[y==1,0], x[y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(x[y==2,0], x[y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(x[y==3,0], x[y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(x[y==4,0], x[y==4,1], s=50, c='blue', label='Cluster 5')

#plot the centroid
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='Centroids')

plt.title('Customer Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
